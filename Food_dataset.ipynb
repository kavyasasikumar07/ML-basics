```python

data = pd.read_csv('/content/food.csv')
age_data = data['Age']
plt.figure(figsize=(8, 6))
plt.hist(age_data, bins=10, edgecolor='black', alpha=0.7)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
data = pd.read_csv('/content/food.csv')
age_data = data['Age']
other_data = data['Nationality']
plt.figure(figsize=(8, 6))
plt.scatter(age_data, other_data, alpha=0.5)
plt.title('Scatter Plot: Age vs. Nationality')
plt.xlabel('Age')
plt.ylabel('Nationality')
plt.show()
```

```python
data = pd.read_csv('/content/food.csv')
a = data['Food']
b = data['Juice']
plt.figure(figsize=(8, 6))
plt.stackplot(data.index, a, b, labels=['a', 'b'])
plt.title('Food and Juice')
plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.legend(loc='upper left')
plt.show()
```

```python
data = pd.read_csv('/content/food.csv')
category_counts = data['Food'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Food')
plt.show()
```

```python
Gender = data['Gender']
Nationality = data['Nationality']
counts = data[data['Age'] > 30]['Gender'].value_counts()
print("Number of males and females above the age 30:")
print(counts)
counts.plot(kind='box', color='c')
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Number of Males and Females above the age 30")
plt.show()
```
